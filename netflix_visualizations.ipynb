### PROBING THE DATA

```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
#in Yahoo Data, Adj Close represents the adjusted close price adjusted for both dividends and splits.
#this means this is the true closing stock price for a given business day.
netflix_stocks = pd.read_csv('NFLX.csv')
netflix_stocks.rename(columns = {'Adj Close':'Price'}, inplace = True)
print(netflix_stocks)
```

```python
dowjones_stocks = pd.read_csv('DJI.csv')
dowjones_stocks.rename(columns = {'Adj Close':'Price'}, inplace = True)
print(dowjones_stocks)
```

```python
#daily data minus weekends and public and bank holidays
netflix_stocks_quarterly = pd.read_csv('NFLX_daily_by_quarter.csv')
netflix_stocks_quarterly.rename(columns = {'Adj Close':'Price'}, inplace = True)

print(netflix_stocks_quarterly.head(), '\n\n')
print(netflix_stocks_quarterly.tail(), '\n\n')
print(netflix_stocks_quarterly.dtypes, '\n\n')
print(len(netflix_stocks_quarterly))
```

### VISUALISING NETFLIX QUARTERLY DATA

The aim is to get an understanding of the distribution of Netflix quarterly stock prices for 2017. Specifically, to see in which quarter stock prices flucutated the most.

```python
import matplotlib.ticker as mtick

def netflix_quarterly_violinplot():
    f, ax = plt.subplots(figsize = (15, 10))
    
    sns.set_style('white')
    
    sns.violinplot(x = 'Quarter', y = 'Price', data = netflix_stocks_quarterly,\
                  inner = 'point', palette = ['#fb6a4a', '#ef3b2c', '#cb181d', '#a50f15'],\
                  linewidth = 0.5)
    
    sns.despine(bottom = True, left = True)
    
    plt.title('Distribution of 2017 Netflix Stock Prices by Quarter', color = '#4f4e4e')
    plt.xlabel('')
    plt.ylabel('')
    
    plt.xticks(fontsize = 12, color = '#4f4e4e')
    plt.yticks([y * 10 for y in range(12, 22)])
    tick = mtick.StrMethodFormatter('${x:,.0f}')
    ax.yaxis.set_major_formatter(tick)
    
    plt.figtext(0.05, 0.0, 'Figure 1.1', fontsize = 12, fontstyle = 'oblique', color = '#4f4e4e')
    
    plt.show()
    
netflix_quarterly_violinplot()
```

```python
print('Minimum Price: ', netflix_stocks_quarterly.Price.min())
print('Maximum Price: ', netflix_stocks_quarterly.Price.max())
print('Yearly IQR: ', netflix_stocks_quarterly.Price.quantile(0.75)\
      - netflix_stocks_quarterly.Price.quantile(0.25), '\n\n')

print('Q1 IQR: ', netflix_stocks_quarterly[netflix_stocks_quarterly.Quarter == 'Q1'].Price.quantile(0.75)\
      - netflix_stocks_quarterly[netflix_stocks_quarterly.Quarter == 'Q1'].Price.quantile(0.25))
print('Q1 Median: ', netflix_stocks_quarterly[netflix_stocks_quarterly.Quarter == 'Q1'].Price.median(), '\n\n')

print('Q2 IQR: ', netflix_stocks_quarterly[netflix_stocks_quarterly.Quarter == 'Q2'].Price.quantile(0.75)\
      - netflix_stocks_quarterly[netflix_stocks_quarterly.Quarter == 'Q2'].Price.quantile(0.25))
print('Q2 Median: ', netflix_stocks_quarterly[netflix_stocks_quarterly.Quarter == 'Q2'].Price.median(), '\n\n')

print('Q3 IQR: ', netflix_stocks_quarterly[netflix_stocks_quarterly.Quarter == 'Q3'].Price.quantile(0.75)\
      - netflix_stocks_quarterly[netflix_stocks_quarterly.Quarter == 'Q3'].Price.quantile(0.25))
print('Q3 Median: ', netflix_stocks_quarterly[netflix_stocks_quarterly.Quarter == 'Q3'].Price.median(), '\n\n')

print('Q4 IQR: ', netflix_stocks_quarterly[netflix_stocks_quarterly.Quarter == 'Q4'].Price.quantile(0.75)\
      - netflix_stocks_quarterly[netflix_stocks_quarterly.Quarter == 'Q4'].Price.quantile(0.25))
print('Q4 Median: ', netflix_stocks_quarterly[netflix_stocks_quarterly.Quarter == 'Q4'].Price.median())
```

- Both median and average Price grows steadily with each Quarter
- Q3 distribution is most spread out

```python
#scatter graph of performance of earnings per share (EPS) using the estimate that
#Yahoo projected for the Quarter compared to the actual earnings for that Quarter.

def eps_scatter_graph():
    f, ax = plt.subplots(figsize = (10, 7))
    plt.xlim(-0.5, 3.5)
    
    sns.set_style('white')
    
    labels = ["1Q2017", "2Q2017", "3Q2017", "4Q2017"]
    earnings_actual =[.4, .15, .29, .41]
    earnings_estimate = [.37, .15, .32, .41 ]

    plt.scatter(x = range(4), y = earnings_actual, marker = 'v', color = 'darkred', alpha = 0.7)
    plt.scatter(x = range(4), y = earnings_estimate, marker = '^', color = 'darkblue', alpha = 0.7)
    
    #sns.despine(bottom = True, left = True)
    
    plt.xticks(range(4), labels, color = '#4f4e4e')
    tick = mtick.StrMethodFormatter('${x:,.2f}')
    ax.yaxis.set_major_formatter(tick)
    
    plt.figtext(0.05, 0.0, 'Figure 2.1', fontsize = 12, fontstyle = 'oblique', color = '#4f4e4e')

    plt.xlabel('')
    plt.ylabel('EPS/ cents\n')
    plt.title('Earnings Per Share (EPS) in Cents')
    plt.legend(['Actual', 'Estimate'], loc = 5)
    
    plt.show()

eps_scatter_graph()
```

```python
#visualising the earnings and revenue reported by Netflix by mapping two bars side-by-side

# The metrics below are in billions of dollars
revenue_by_quarter = [2.79, 2.98, 3.29, 3.7]
earnings_by_quarter = [.0656, .12959, .18552, .29012]
labels = ["2Q2017", "3Q2017", "4Q2017", "1Q2018"]

df = pd.DataFrame({'Amount': revenue_by_quarter + earnings_by_quarter, 'Revenue_or_earnings': \
                   ['Revenue'] * 4 + ['Earnings'] * 4, 'Quarter': labels * 2})
print(df)

def revenue_earnings_barplot():
    f, ax = plt.subplots(figsize = (11, 7))
    
    sns.set_style('white')
    
    sns.barplot(x = 'Quarter', y = 'Amount', hue = 'Revenue_or_earnings', data = df,\
               palette = ['darkgreen', 'darkred'], ci = None)
    
    sns.despine(bottom = True, left = True)
    
    plt.title('Netflix Revenue and Reported Earnings per Quarter                           ')
    plt.xlabel('')
    plt.ylabel('')
    
    plt.xticks(fontsize = 12, color = '#4f4e4e')
    plt.yticks([], [])
    
    plt.legend(title = '', labels = ['Revenue', 'Earnings'], loc = 2)
    plt.xlim(-0.42, 4)
    
    for p in ax.patches:
        height = p.get_height()
        ax.text(x = p.get_x() + p.get_width() / 2, y = height + 0.025,\
               s = '$ {:.5f}'.format(height), ha = 'center', color = '#4f4e4e', fontsize = 11)
    
    plt.figtext(0.05, 0.0, 'Figure 3.1', fontsize = 12, fontstyle = 'oblique')

revenue_earnings_barplot()
```

```python
for i in range(4):
    print(round((df.Amount[i + 4] / df.Amount[i]) * 100, 2))
```

- Both, revenue and earnings follow an upward trend;
- Earnings constitute between 2% and 8% of revenue for each quarter. This % increases with each quarter.

```python
def netflix_dowjones_stock_price_trend():
    plt.figure(figsize = (15, 10))
    
    sns.set_style('white')

    #left plot Netflix
    ax1 = plt.subplot(1, 2, 1)
    sns.lineplot(x = netflix_stocks['Date'], y = netflix_stocks['Price'], color = 'darkred')
    plt.ylim(140, 200)
    ax1.set_title('Netflix')
    ax1.set_xlabel('Month')
    ax1.set_ylabel('Stock Price')
    plt.xticks(range(12), labels = range(1, 13), fontsize = 12, color = '#4f4e4e')
    plt.yticks([y * 10 for y in range(14, 21)], fontsize = 12, color = '#4f4e4e')
    tick = mtick.StrMethodFormatter('${x:,.0f}')
    ax1.yaxis.set_major_formatter(tick)

    #right plot Dow Jones
    ax2 = plt.subplot(1, 2, 2)
    sns.lineplot(x = dowjones_stocks['Date'], y = dowjones_stocks['Price'], color = 'darkblue')
    plt.ylim(19800, 25000)
    ax2.set_title('Dow Jones')
    ax2.set_xlabel('Month')
    ax2.set_ylabel('Stock Price')
    plt.xticks(range(12), labels = range(1, 13), fontsize = 12, color = '#4f4e4e')
    plt.yticks([y * 1000 for y in range(19, 26)], fontsize = 12, color = '#4f4e4e')
    tick = mtick.StrMethodFormatter('${x:,.0f}')
    ax2.yaxis.set_major_formatter(tick)

    plt.subplots_adjust(wspace = 0.25)
    
    plt.figtext(0.09, 0.0, 'Figure 4.1', fontsize = 12, fontstyle = 'oblique')

netflix_dowjones_stock_price_trend()
```

- Netflix followed an upward trend in 2017 like the DJ Industrial Average but was more volatile;
- Netflix stock was more volatile, with two sharp dips in Month 6 and Month 8, while the same is not observed in Dow Jones;
- The average stock price of Netflix is low compared to DJ Industrial Average.
